```python
plt.figure(figsize=(12,8))
sns.countplot(data=sasi,x="ship-state")
plt.show()
```

```python
top_10_states = sasi['ship-state'].value_counts().head(10).index

# Filter the DataFrame to include only top 10 states
filtered_data = sasi[sasi['ship-state'].isin(top_10_states)]

# Plot the count plot for top 10 states
plt.figure(figsize=(12, 8))
sns.countplot(data=filtered_data, x="ship-state", order=top_10_states)
plt.title("Top 10 States by Shipment Count")
plt.xlabel("State")
plt.ylabel("Count")
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.show()
```
